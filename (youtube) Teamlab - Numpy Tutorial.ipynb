### Groupby - grouped

1. Aggregation : 요약된 통계정보를 추출함
2. Transformation: 해당 정보를 변환해줌.
3. Filtration: 특정 정보를 제거해 보여주는 필터 기능.

```python
grouped.agg(np.mean)
```

```python
grouped['Points'].agg([np.sum, np.mean, np.std])
```

 ### Transformation

```python
score = lambda x : (x - x.mean()) / x.std()
grouped.transform(score)
```

## Feature Scaling

```python
df = pd.io.parsers.read_csv(
    'https://raw.githubusercontent.com/rasbt/pattern_classification/master/data/wine_data.csv',
header = None, usecols = [0,1,2])

df.columns = ['Class label', 'Alcohol', 'Malic acid']
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

std_scale = StandardScaler().fit(df[['Alcohol', 'Malic acid']]) # fit을 통해 규칙 생성 후 저장
df_std = std_scale.transform(df[['Alcohol', 'Malic acid']]) # transform을 통해 변형

minmax_scale = preprocessing.MinMaxScaler().fit(df[['Alcohol', 'Malic acid']])
df_minmax = minmax_scale.transform(df[['Alcohol', 'Malic acid']])
```
